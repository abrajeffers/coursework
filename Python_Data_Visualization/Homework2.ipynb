```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
weather = pd.read_csv("/Users/abrajeffers/Downloads/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv")
id_leapyear = weather['Date'].str.contains('02-29')
weather = weather[~id_leapyear]
weather['Data_Value']= weather['Data_Value']/10
id_2004 = weather['Date'].str.contains('2004')
weather = weather[-id_2004]
id_2015 = weather['Date'].str.contains('2015')
df_2015 = weather[id_2015]
weather = weather[-id_2015]
#####MAKE COLUMNS FOR DATE AND YEAR
weather['Month'] = weather['Date'].apply(lambda x: x[5:7])
weather['Day'] = weather['Date'].apply(lambda x: x[8:10])
df_2015['Month'] = df_2015['Date'].apply(lambda x: x[5:7])
df_2015['Day'] = df_2015['Date'].apply(lambda x: x[8:10])
id_tmax = weather['Element'].apply(lambda x: x=='TMAX')
df_min = weather[-id_tmax]
df_min = df_min.groupby(by=['Month', 'Day']).min()
id_tmin = weather['Element'].apply(lambda x: x=='TMIN')
df_max = weather[-id_tmin]
df_max = df_max.groupby(by=['Month', 'Day']).max()
id_tmax2015 = df_2015['Element'].apply(lambda x: x=='TMAX')
df_max2015 = df_2015[id_tmax2015]
df_max2015 = df_max2015.groupby(by=['Month', 'Day']).max()
id_tmin2015 = df_2015['Element'].apply(lambda x: x=='TMIN')
df_min2015 = df_2015[id_tmin2015]
df_min2015 = df_min2015.groupby(by=['Month', 'Day']).max()
df_max2015
```

```python
pivot_min = pd.pivot_table(df_min, columns = ['Element'], index = ['Month','Day'], values = ['Data_Value'])
pivot_max = pd.pivot_table(df_max, columns = ['Element'], index = ['Month','Day'], values = ['Data_Value'])
pivot_min2015 = pd.pivot_table(df_min2015, columns = ['Element'], index = ['Month','Day'], values = ['Data_Value'])
pivot_max2015 = pd.pivot_table(df_max2015, columns = ['Element'], index = ['Month','Day'], values = ['Data_Value'])
pivot_min
```

```python
pivot_min.columns = pivot_min.columns.droplevel()
pivot_max.columns = pivot_max.columns.droplevel()
pivot_min2015.columns = pivot_min2015.columns.droplevel()
pivot_max2015.columns = pivot_max2015.columns.droplevel()
pivot_min2015
```

```python
df_full = pivot_max.merge(pivot_min, how = 'outer', left_on = ['Month', 'Day'], right_on = ['Month', 'Day'])
df_full =  df_full.reset_index(level=0, drop=True)
df_full.index = range(0,365)
df_full2015 = pivot_max2015.merge(pivot_min2015,how='outer',left_on=['Month','Day'],right_on=['Month','Day'])
df_full2015 =  df_full2015.reset_index(level=0, drop=True)
df_full2015.index = range(0,365)
for i, row in df_full2015.iterrows():
    if row['TMAX'] <= df_full.iloc[i]['TMAX']:
        row['TMAX'] = np.nan
    if row['TMIN'] >= df_full.iloc[i]['TMIN']:
        row['TMIN'] = np.nan
df_full2015
```

```python
#Plot line graph
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
import matplotlib.dates
from matplotlib import lines


# create a new figure
fig = Figure()

# associate fig with the backend
canvas = FigureCanvasAgg(fig)

# add a subplot to the fig
ax = fig.add_subplot(111)


# create a new figure
plt.figure()

x = range(0,df_full.shape[0])
y20052014max = df_full['TMAX']
y20052014min = df_full['TMIN']
y2015max = df_full2015['TMAX']
y2015min = df_full2015['TMIN']

plt.plot(x,y20052014max, color='green')
plt.plot(x,y20052014min, color = "skyblue")
plt.scatter(x,y2015max, color = 'red', s= 12)
plt.scatter(x,y2015min, color = "blue", s = 12)


# get the current axes
ax = plt.gca()


# Set axis properties [xmin, xmax, ymin, ymax]
ax.axis([0,365,-40,50])
xticks = (pd.date_range('1/1/2015',freq = 'M',periods=12)- 1 + pd.Timedelta('1D')).strftime("%-j").astype(int)
ax.set_xticks(xticks)
xticks_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
ax.set_xticklabels(xticks_labels)
plt.xlabel('Month')
plt.ylabel('Temperature (Celcius)')
plt.title('Record Highs and Lows in Ann Arbor, MI, 2005-2015')
plt.legend(['Record High 2005-2014', 'Record Low 2005-2014', 'High of 2015 beyond 2005-2014', 'Low of 2015 beyond 2005-2014'])
ax.fill_between(x,y20052014max,y20052014min,color='grey')






# save the figure to test.png
plt.savefig('HW2fig.png')

```
